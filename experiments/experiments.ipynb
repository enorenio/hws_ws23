```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_pdf(x, mu, rho):
    return 1 / (np.sqrt(2 * np.pi * rho**2)) * np.exp(-(x - mu)**2 / (2 * rho**2))

def alt_gaussian_pdf(x, mu, rho):
    return 1 / (np.sqrt(2 * np.pi * rho**2)) * np.exp(-(x - mu)**2 / (2))

# Define parameters
mu = 0  # mean
rho = 3  # standard deviation

# Generate x values
x_values = np.linspace(-5, 5, 1000)

# Calculate corresponding y values using the Gaussian PDF
y_gauss = gaussian_pdf(x_values, mu, rho)
y_alt = alt_gaussian_pdf(x_values, mu, rho)

# Plot the function
plt.plot(x_values, y_gauss, label=f'Gauss μ={mu}, ρ={rho}')
plt.plot(x_values, y_alt, label=f'Alt μ={mu}, ρ={rho}')
plt.title('Comparison Alt and Gaussian Probability Density Function')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
